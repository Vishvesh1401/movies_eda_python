```python
n_data['orig_lang'] =  n_data['orig_lang'].str.strip()
```

```python
n_data.query('orig_lang in "Japanese"')
```

```python
def category(row):
    if row['score'] >= 90:
        return 'Blockbuster'
    elif row['score'] >= 70 and row['score'] < 90:
        return "Hit"
    else:
        return "Average"

n_data.loc[:,"Category"] = n_data.apply(category, axis=1)
```

```python
n_data.head()
```

```python
n_data = n_data.drop('Category', axis=1)
```

```python
n_data.head()
```

```python
condition = [
    (n_data['score'] >= 90),
    (n_data['score'] >= 70) & (n_data['score'] < 90),
    (n_data['score'] < 70)  
]

choices = ["Blockbuster", "Hit", "Average"]

n_data["Category"] = np.select(condition, choices, default="Unknown")
```
